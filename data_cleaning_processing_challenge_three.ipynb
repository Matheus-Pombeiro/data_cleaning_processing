# **Pandas: limpeza e tratamento de dados - Desafio 3**

## Utilizando dados duplicados e nulos

```python
# Importando bibliotecas necessárias
import pandas as pd
import json
```

```python
# Lendo a base de dados
dados = pd.read_json('/content/cursos_cadastrados.json')
dados
```

```python
# Abrindo o dataframe
with open('/content/cursos_cadastrados.json', 'r') as f:
  json_bruto = json.load(f)
```

```python
# Normalizando o dataframe
dados = pd.json_normalize(json_bruto)
dados
```

```python
# Somando os dados nulos no dataframe
dados.isnull().sum().sum()
```

```python
# Verificando onde estão os valores nulos
dados[dados.isna().any(axis=1)]
```

```python
# Removendo dados nulos
dados.dropna(inplace=True)
```

```python
# Verificando novamente a quatidade dados nulos
dados.isnull().sum().sum()
```

```python
# Verificando a presença de dados duplicados
dados.duplicated()
```

```python
# Removendo dados duplicados
dados.drop_duplicates(inplace=True)
```

```python
# Verificando a quantidade de dados duplicados
dados.duplicated().sum()
```

```python
''' Existe mais de uma coluna com strings vazias como dados,
por isso vamos substituir essas strings por dados nulos '''
dados.replace('', pd.NA, inplace=True)
dados
```

```python
# Removendo os dados nulos
dados.dropna(inplace=True)
```

```python
# Resetando o index
dados.reset_index(drop=True)
```

```python
# Verificando o tipo dos dados
dados.info()
```

```python
# Convertendo tipos de dados
dados['concluintes'] = dados['concluintes'].astype(int)
dados['data_inicio'] = pd.to_datetime(dados['data_inicio'])
dados['data_conclusao'] = pd.to_datetime(dados['data_conclusao'])
dados['preco'] = dados['preco'].astype(float)
```
