## Conhecendo os dados

```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import json
```

```python
# Lendo o dataset
dados = pd.read_json('/content/informacoes.json')
dados.head()
```

```python
# Abrindo o arquivo
with open('/content/informacoes.json') as f:
  dados = json.load(f)
```

```python
# Normalizando os dados
dados_normalizados = pd.json_normalize(dados, record_path='enderecos', meta=['nome', 'idade'])
dados_normalizados.head()
```
