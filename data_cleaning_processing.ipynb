# **Pandas: limpeza e tratamento de dados**

## Conhecendo os dados

```python
# Importando bibliotecas
import pandas as pd
import numpy as np
```

```python
# Lendo o dataset
dados_churn = pd.read_json('/content/dataset-telecon.json')
dados_churn.head(3)
```

```python
# Verificando dados da coluna 'conta'
dados_churn['conta'][0]
```

```python
# Normalizando os dados de 'telefone'
pd.json_normalize(dados_churn['telefone']).head(3)
```

```python
# Importando a biblioteca json
import json
```

```python
# Abrindo o arquivo
with open('/content/dataset-telecon.json') as f:
  json_bruto = json.load(f)
```

```python
# Normalizando os dados
dados_normalizados = pd.json_normalize(json_bruto)
dados_normalizados.head(3)
```

## Transformação inicial dos dados

```python
# Buscando informações sobre o dataframe
dados_normalizados.info()
```

```python
# Filtrando os dados
dados_normalizados[dados_normalizados['conta.cobranca.Total'] == ' '].head(3)
```

```python
# Filtrando e selecionando dados
dados_normalizados[dados_normalizados['conta.cobranca.Total'] == ' '][['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

```python
# Encontrando os índices
idx = dados_normalizados[dados_normalizados['conta.cobranca.Total'] == ' '].index
```

```python
# Passando valores para a coluna
dados_normalizados.loc[idx, 'conta.cobranca.Total'] = dados_normalizados.loc[idx, 'conta.cobranca.mensal'] * 24
```

```python
# Passabndo a quantidade de meses para a coluna
dados_normalizados.loc[idx, 'cliente.tempo_servico'] = 24
```

```python
# Encontrando colunas baseadi no index
dados_normalizados.loc[idx, ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

```python
# Convertendo o tipo dos dados
dados_normalizados['conta.cobranca.Total'] = dados_normalizados['conta.cobranca.Total'].astype(np.float64)
```

```python
# Conferindo os dados
dados_normalizados.info()
```

```python
# Indentificando valores únicos presente nas colunas
for col in dados_normalizados.columns:
  print(f'Coluna: {col}')
  print(dados_normalizados[col].unique())
  print('-' * 30)
```

```python
# Fitrando os dados de churn
dados_normalizados.query('`Churn` == ""')
```

```python
# Fazendo uma cópia independente do dataframe com copy()
dados_sem_vazio = dados_normalizados[dados_normalizados['Churn'] != ''].copy()
```

```python
# Verificando os ídices do novo dataframe
dados_sem_vazio.info()
```

```python
# Resetando os índices
dados_sem_vazio.reset_index(drop=True, inplace=True)
```

```python
# Verificando os novos índices do dataframe
dados_sem_vazio.info()
```

## Utilizando dados duplicados e nulos

```python
# Verificando dados duplicados
dados_sem_vazio.duplicated()
```

```python
# Somando o número de dados duplicados
dados_sem_vazio.duplicated().sum()
```

```python
# Recebendo o dados duplicados
filtro_duplicadas = dados_sem_vazio.duplicated()
```

```python
# Verificando os registros duplicados no dataframe
dados_sem_vazio[filtro_duplicadas]
```

```python
# Removendo os dados duplicados
dados_sem_vazio.drop_duplicates(inplace=True)
```

```python
# Verificando a existêcias de dados duplicados
dados_sem_vazio.duplicated().sum()
```

```python
# Verificando se apenas os dados duplicados foram removidos do dataframe
dados_sem_vazio.shape
```

```python
# Verificando a existência de dados nulos
dados_sem_vazio.isna()
```

```python
# Somando a quantidade de dados nulos por coluna
dados_sem_vazio.isna().sum()
```

```python
# Somando o total de dados nulos no dataframe
dados_sem_vazio.isna().sum().sum()
```

```python
# Filtrando os dados nulos. Cada linha tem pelo menos 1 valor nulo (any)
dados_sem_vazio[dados_sem_vazio.isna().any(axis=1)]
```

```python
# Recebendo os dados nulos de uma coluna
filtro = dados_sem_vazio['cliente.tempo_servico'].isna()
```

```python
# Verificando as linhas com dados nulos das colunas
dados_sem_vazio[filtro][['cliente.tempo_servico', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

```python
dados_sem_vazio['cliente.tempo_servico'] = dados_sem_vazio['cliente.tempo_servico'].fillna(
    np.ceil(
        dados_sem_vazio['conta.cobranca.Total'] / dados_sem_vazio['conta.cobranca.mensal']
    )
)
```

```python
# Verificando as colunas que continham dados nulos
dados_sem_vazio[filtro][['cliente.tempo_servico', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

```python
# Verificando os dados nulos na coluna
dados_sem_vazio['cliente.tempo_servico'].isna().sum()
```

```python
# Contado cada valor da coluna
dados_sem_vazio['conta.contrato'].value_counts()
```

```python
# Recebendo as colunas que devem ser dropadas
colunas_dropar = ['conta.contrato', 'conta.faturamente_eletronico', 'conta.metodo_pagamento']
```

```python
# Somando os dados vazios nas colunas selecionadas
dados_sem_vazio[colunas_dropar].isna().any(axis=1).sum()
```

```python
# Removendo dados nulos das colunas selecionadas e copiando para um novo dataframe
df_sem_nulo = dados_sem_vazio.dropna(subset=colunas_dropar).copy()
df_sem_nulo
```

```python
# Resetando o índice
df_sem_nulo.reset_index(drop=True, inplace=True)
```

```python
# Verificando a quantidade de dados nulos por coluna
df_sem_nulo.isna().sum()
```
