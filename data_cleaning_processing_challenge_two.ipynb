# **Pandas: limpeza e tratamento de dados - Desafio 2**

## Initial data transformation

```python
# Importando bibliotecas necessárias
import pandas as pd
import json
```

```python
# Lendo os dados
dados = pd.read_json('/content/pessoas.json')
dados
```

```python
# Normalizando os dados - Abrindo os dados
with open('/content/pessoas.json', 'r') as f:
  json_bruto = json.load(f)
```

```python
# Normalizando os dados - Recebendo em um dataframe
dados_normalizados = pd.json_normalize(json_bruto, record_path='pessoas')
dados_normalizados
```

```python
# Buscando informações sobre o dataframe
dados_normalizados.info()
```

```python
# Convertendo o tipo dos dados de idade
dados_normalizados['idade'] = dados_normalizados['idade'].astype(int)
```

```python
# Verificando a converção
dados_normalizados.info()
```

```python
# Filtrando os dados e separando eles
dados_filtrados = dados_normalizados.query('`endereco.rua` != ""').copy()
dados_filtrados
```
